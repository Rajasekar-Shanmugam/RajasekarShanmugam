# Stage 1 (Data understanding, preparation, and feature engineering)

# Data understanding
All important data quality checks are performed and inconsistent/missing data is handled appropriately.

```python
# reading the input data and preview
churn_original= pd.read_csv('telecom_churn_data.csv')

churn= churn_original.copy()
churn.head()
# Understand number of rows and columns and sneak of first 5 row
print (churn.shape); 
churn.info()
churn.head(5)
```

```python
# check the % count of null
round(100*(churn.isnull().sum()/len(churn.index)), 2).sort_values(ascending=False).head(50)
```

### Handling missing values , Impute missing values

```python
# Dropping columns mobile and circle as they do not contribue to model building
id_cols = ['mobile_number', 'circle_id']
churn.drop(id_cols, axis=1, inplace=True)
```

```python
# list of columns to impute missing values with zeroes
recharge_col = ['total_rech_data_6', 'total_rech_data_7', 'total_rech_data_8', 'total_rech_data_9',
            'av_rech_amt_data_6', 'av_rech_amt_data_7', 'av_rech_amt_data_8', 'av_rech_amt_data_9',
            'arpu_2g_6', 'arpu_2g_7', 'arpu_2g_8', 'arpu_2g_9',
            'arpu_3g_6', 'arpu_3g_7', 'arpu_3g_8', 'arpu_3g_9',
            'count_rech_2g_6', 'count_rech_2g_7', 'count_rech_2g_8', 'count_rech_2g_9', 
            'count_rech_3g_6', 'count_rech_3g_7', 'count_rech_3g_8', 'count_rech_3g_9', 
            'max_rech_data_6', 'max_rech_data_7', 'max_rech_data_8', 'max_rech_data_9',
            'total_rech_data_6', 'total_rech_data_7', 'total_rech_data_8', 'total_rech_data_9', 
            'fb_user_6', 'fb_user_7', 'fb_user_8', 'fb_user_9', 
            'night_pck_user_6', 'night_pck_user_7', 'night_pck_user_8', 'night_pck_user_9']
# impute missing values with 0
churn[recharge_col] = churn[recharge_col].apply(lambda x: x.fillna(0))
```

```python
# check the % count of null again
round(100*(churn.isnull().sum()/len(churn.index)), 2).sort_values(ascending=False).head(10)
```

# Check for Categorical Variables and Create Dummy variables


```python
# Let’s impute missing value with  '-1' 

catg_vars = ["night_pck_user_6", "night_pck_user_7", "night_pck_user_8",  "night_pck_user_9", "fb_user_6", "fb_user_7", "fb_user_8", "fb_user_9"]

for var in catg_vars:
    churn[var].fillna(value=-1, inplace=True)

# Create dummy variables for all the six categorial variables

night_6_dummy= pd.get_dummies(churn['night_pck_user_6'],prefix='night_6', drop_first=True)
night_7_dummy= pd.get_dummies(churn['night_pck_user_7'],prefix='night_7', drop_first=True)
night_8_dummy= pd.get_dummies(churn['night_pck_user_8'],prefix='night_8', drop_first=True)
night_9_dummy= pd.get_dummies(churn['night_pck_user_9'],prefix='night_9', drop_first=True)


fb_6_dummy = pd.get_dummies(churn['fb_user_6'],prefix='fb_6', drop_first=True)
fb_7_dummy = pd.get_dummies(churn['fb_user_7'],prefix='fb_7', drop_first=True)
fb_8_dummy = pd.get_dummies(churn['fb_user_8'],prefix='fb_8', drop_first=True)
fb_9_dummy = pd.get_dummies(churn['fb_user_9'],prefix='fb_9', drop_first=True)


churn = pd.concat([churn, night_6_dummy], axis=1)
churn = pd.concat([churn, night_7_dummy], axis=1)
churn = pd.concat([churn, night_8_dummy], axis=1)
churn = pd.concat([churn, night_9_dummy], axis=1)


churn = pd.concat([churn, fb_6_dummy], axis=1)
churn = pd.concat([churn, fb_7_dummy], axis=1)
churn = pd.concat([churn, fb_8_dummy], axis=1)
churn = pd.concat([churn, fb_9_dummy], axis=1)


# drop all categorial columns
churn = churn.drop(['night_pck_user_6','night_pck_user_7','night_pck_user_8','night_pck_user_9','fb_user_6','fb_user_7','fb_user_8','fb_user_9'], 1)
print (churn.shape); 
churn.info()
churn.head(5)
```

```python
# covert date columns to datetime format
date_vars = ["date_of_last_rech_6",  "date_of_last_rech_7", "date_of_last_rech_8","date_of_last_rech_9"]
churn[date_vars].head()

# convert to datetime
for col in date_vars:
    churn[col] = pd.to_datetime(churn[col])

print(churn[date_vars].info())
churn[date_vars].head()

```

```python
# Create new days columns, instead of date
import datetime
last_date_of_month_6 = datetime.datetime.strptime("30-06-2014", "%d-%m-%Y")
last_date_of_month_7 = datetime.datetime.strptime("31-07-2014", "%d-%m-%Y")
last_date_of_month_8 = datetime.datetime.strptime("31-08-2014", "%d-%m-%Y")
last_date_of_month_9 = datetime.datetime.strptime("31-08-2014", "%d-%m-%Y")


churn["rech_days_left_6"]      = (last_date_of_month_6 - churn.date_of_last_rech_6).astype('timedelta64[D]')
churn["rech_days_left_7"]      = (last_date_of_month_7 - churn.date_of_last_rech_7).astype('timedelta64[D]')
churn["rech_days_left_8"]      = (last_date_of_month_8 - churn.date_of_last_rech_8).astype('timedelta64[D]')
churn["rech_days_left_9"]      = (last_date_of_month_9 - churn.date_of_last_rech_9).astype('timedelta64[D]')


day_columns = ["rech_days_left_6", "rech_days_left_7", "rech_days_left_8","rech_days_left_9"]

print(churn[day_columns].head(10))
print(churn[day_columns].info())

```

```python
# Drop all old date columns: add dates columns to drop_column list
churn.drop(date_vars, axis=1, inplace=True)
churn = churn.copy()

print('Dataframe Shape: ', churn.shape)
```
